### 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.scatterplot(x=df['condition'], y = df['price']);
```

Согласно этого графика, можно заметить, что на недвижимость с состоянием 3, цена чаще всего выше

```python
sns.jointplot(x=df['bedrooms'], y=df['price'], kind='reg');
```

В этом графике, можно увидеть, что кол-во комнат имеет значение при стоимости недвижимости

```python
sns.boxplot(x=df['waterfront'], y=df['price'], whis=1.5)
```

В данном графике, можно обратить внимание, что средняя стоимость недвижиости выше с видом на набережную

```python
sns.jointplot(x=df['grade'], y=df['price'], kind='kde');
```

По этому графику, можем сделать вывод, что чем выше оценка недвижимости, тем выше стоимость. Так же видно, что в основном недвижимость имеет оценуку от 6 до 8

```python
corr_matrix = df.corr().round(2)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
plt.figure(figsize=(10, 8))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

По данной матрице, мы можем сделать вывод, что на стоимость недвижимости в большей степении влияют следующие параметры:
1) Жилая площадь
2) Оценка недвижимости
3) Кол-во вынных комнат
4) Вид из окна
5) Наличие подвала
6) Этаж
7) Набережная
